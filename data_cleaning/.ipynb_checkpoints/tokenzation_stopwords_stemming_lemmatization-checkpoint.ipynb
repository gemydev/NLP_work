```python
import pandas as pd
import re
```

```python
data = pd.read_csv("../data/amazon_alexa.tsv",delimiter="\t")
```

```python
def special_char_removal(text):
    pat = r'^[a-zA-Z0-9]'
    return re.sub(pat,' ',text)

data['verified_reviews'] = data.apply(lambda x:  special_char_removal(x['verified_reviews']),axis=1)
```

**Tokenzation**

```python
# for tokenzation
import nltk
from nltk.tokenize import sent_tokenize , word_tokenize
```

```python
# sentence tokenization
data["sent_tokenization"] = data.apply(lambda x : sent_tokenize(x["verified_reviews"]), axis=1)
data
```

```python
# Word tokenization
data["word_tokenization"] = data.apply(lambda x : word_tokenize(x["verified_reviews"]), axis=1)
data
```

**Stop Words in nltk**

```python
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
def stopwords_removal(list_text):
    list_text_new = []
    for i in list_text:
        if not i in stopwords.words():
            list_text_new.append(i)
    return ' '.join(list_text_new)
```

```python
# let's remove the stop words
data["sent_without_stop_words(nltk)"] = data.apply(lambda x : stopwords_removal(x["word_tokenization"]), axis=1)
data
```

**Stemming (using nltk)**

```python
# function for stemming
def get_stem(text):
    stemmer = nltk.porter.PorterStemmer()
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text

data["stemming"] = data.apply(lambda x : get_stem(x["verified_reviews"]), axis=1)
data
```

**Lemmatization (using spacy)**

```python
# pip3 install https://github.com/explosion/spacy-models/releases/download/en_core_web_sm-2.2.0/en_core_web_sm-2.2.0.tar.gz
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# function for lemmatization
def get_lem(text):
    text = nlp(text)
    lemmatized_sentence = " ".join([word.lemma_ for word in text])
    return lemmatized_sentence

data["stemming"] = data.apply(lambda x : get_lem(x["verified_reviews"]), axis=1)
data
```
